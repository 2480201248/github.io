```python
import pandas as pd
import json

# 加载Excel文件
file_path = '../attachment/1596531496369353123.xlsx'
excel_data = pd.read_excel(file_path, sheet_name=None)

# 读取“采购预测”工作簿
purchase_forecast_data = excel_data['采购预测']
purchase_forecast_data['月份'] = purchase_forecast_data['日期'].dt.to_period('M')

# 按月份和物料分类进行分组并汇总实际出库量
grouped_data = purchase_forecast_data.groupby(['月份', '物料分类'])['实际出库'].sum().unstack(fill_value=0)

# 转换数据为适合绘图的格式
stacked_bar_chart_data = grouped_data.reset_index()
months = stacked_bar_chart_data['月份'].astype(str).tolist()
categories = grouped_data.columns.tolist()
series_data = []

for category in categories:
    series_data.append({
        'name': category,
        'type': 'bar',
        'stack': '总量',
        'data': stacked_bar_chart_data[category].tolist()
    })

# 生成JS代码
js_code_chart3 = f"""
// 堆叠柱状图 - Chart3
var chart3 = echarts.init(document.getElementById('chart3'));
var option3 = {{
    title: {{
        text: '按月份展示物料分类的出库汇总'
    }},
    tooltip: {{
        trigger: 'axis',
        axisPointer: {{ type: 'shadow' }}
    }},
    legend: {{
        data: {json.dumps(categories)}
    }},
    grid: {{
        left: '3%',
        right: '4%',
        bottom: '3%',
        containLabel: true
    }},
    xAxis: {{
        type: 'category',
        data: {json.dumps(months)}
    }},
    yAxis: {{
        type: 'value'
    }},
    series: {json.dumps(series_data)}
}};
chart3.setOption(option3);
"""

# 输出JS代码
print(js_code_chart3)
```

```python
import pandas as pd
import json

# 加载Excel文件
file_path = '../attachment/1596531496369353123.xlsx'
excel_data = pd.read_excel(file_path, sheet_name=None)

# 读取“采购计划_2”工作簿
purchase_plan_data = excel_data['采购计划_2']
purchase_plan_data['月份'] = purchase_plan_data['日期'].dt.to_period('M')

# 按月份进行分组并汇总数据
monthly_summary = purchase_plan_data.groupby('月份').agg({
    '预测销售出库数量': 'sum',
    '实际销售出库数量': 'sum',
    '预测误差率': 'mean'
}).reset_index()

# 转换数据为适合绘图的格式
months = monthly_summary['月份'].astype(str).tolist()
forecast_sales = monthly_summary['预测销售出库数量'].tolist()
actual_sales = monthly_summary['实际销售出库数量'].tolist()
error_rate = monthly_summary['预测误差率'].tolist()

# 生成JS代码
js_code_chart1 = f"""
// 组合图表 - Chart1
var chart1 = echarts.init(document.getElementById('chart1'));
var option1 = {{
    title: {{
        text: '每月销售出库数量及预测误差'
    }},
    tooltip: {{
        trigger: 'axis',
        axisPointer: {{ type: 'cross' }}
    }},
    legend: {{
        data: ['预测销售出库数量', '实际销售出库数量', '预测误差率']
    }},
    grid: {{
        left: '3%',
        right: '4%',
        bottom: '3%',
        containLabel: true
    }},
    xAxis: {{
        type: 'category',
        data: {json.dumps(months)}
    }},
    yAxis: [
        {{
            type: 'value',
            name: '销售出库数量'
        }},
        {{
            type: 'value',
            name: '预测误差率',
            axisLabel: {{
                formatter: '{{value}} %'
            }}
        }}
    ],
    series: [
        {{
            name: '预测销售出库数量',
            type: 'bar',
            data: {json.dumps(forecast_sales)}
        }},
        {{
            name: '实际销售出库数量',
            type: 'bar',
            data: {json.dumps(actual_sales)}
        }},
        {{
            name: '预测误差率',
            type: 'line',
            yAxisIndex: 1,
            data: {json.dumps(error_rate)}
        }}
    ]
}};
chart1.setOption(option1);

// 添加定时器逻辑，每隔1秒切换展示数据
var currentIndex = -1;
setInterval(function () {{
    var dataLen = option1.series[0].data.length;
    // 取消之前高亮的图形
    chart1.dispatchAction({{
        type: 'downplay',
        seriesIndex: 0,
        dataIndex: currentIndex
    }});
    currentIndex = (currentIndex + 1) % dataLen;
    // 高亮当前图形
    chart1.dispatchAction({{
        type: 'highlight',
        seriesIndex: 0,
        dataIndex: currentIndex
    }});
    // 显示 tooltip
    chart1.dispatchAction({{
        type: 'showTip',
        seriesIndex: 0,
        dataIndex: currentIndex
    }});
}}, 1000);
"""

# 输出JS代码
print(js_code_chart1)
```

```python
import pandas as pd
import json

# 加载Excel文件
file_path = '../attachment/1596531496369353123.xlsx'
excel_data = pd.read_excel(file_path, sheet_name=None)

# 读取“采购计划_1”工作簿
purchase_plan_1_data = excel_data['采购计划_1']

# 数据预处理
# 柱状图数据：按物料等级计算数量
material_level_counts = purchase_plan_1_data['物料等级'].value_counts().reset_index()
material_level_counts.columns = ['物料等级', '数量']

# 饼图数据：按采购单位计算数量
purchase_unit_counts = purchase_plan_1_data['负责人'].value_counts().reset_index()
purchase_unit_counts.columns = ['负责人', '数量']

# 转换数据为适合绘图的格式
bar_chart_data = {
    'categories': material_level_counts['物料等级'].tolist(),
    'data': material_level_counts['数量'].tolist()
}

pie_chart_data = [{
    'name': row['负责人'],
    'value': row['数量']
} for index, row in purchase_unit_counts.iterrows()]

# 生成柱状图和饼图的JS代码
js_code_chart2_and_chart3 = f"""
// 柱状图 - Chart2
var chart2 = echarts.init(document.getElementById('chart2'));
var option2 = {{
    title: {{
        text: '物料等级分布'
    }},
    tooltip: {{
        trigger: 'axis',
        axisPointer: {{ type: 'shadow' }}
    }},
    xAxis: {{
        type: 'category',
        data: {json.dumps(bar_chart_data['categories'])}
    }},
    yAxis: {{
        type: 'value'
    }},
    series: [{{
        data: {json.dumps(bar_chart_data['data'])},
        type: 'bar'
    }}]
}};
chart2.setOption(option2);

// 饼图 - Chart3
var chart3 = echarts.init(document.getElementById('chart3'));
var option3 = {{
    title: {{
        text: '负责人占比'
    }},
    tooltip: {{
        trigger: 'item',
        formatter: '{{a}} <br/>{{b}}: {{c}} ({{d}}%)'
    }},
    series: [{{
        name: '负责人',
        type: 'pie',
        radius: '50%',
        data: {json.dumps(pie_chart_data)},
        emphasis: {{
            itemStyle: {{
                shadowBlur: 10,
                shadowOffsetX: 0,
                shadowColor: 'rgba(0, 0, 0, 0.5)'
            }}
        }}
    }}]
}};
chart3.setOption(option3);
"""

# 输出JS代码
print(js_code_chart2_and_chart3)

```

```python
import pandas as pd
import json

# 加载Excel文件
file_path = '../attachment/1596531496369353123.xlsx'
excel_data = pd.read_excel(file_path, sheet_name=None)

# 读取“采购库存”工作簿
purchase_inventory_data = excel_data['采购库存']

# 数据预处理：按仓库筛选并聚合数据
xa_data = purchase_inventory_data[purchase_inventory_data['仓库'] == '厦门A仓']
gd_data = purchase_inventory_data[purchase_inventory_data['仓库'] == '广东B仓']
sz_data = purchase_inventory_data[purchase_inventory_data['仓库'] == '深圳C仓']

# 定义一个函数来计算每个仓库的汇总数据（不包括“可用天数”和“缺货天数”）
def calculate_summary(data):
    summary = data.agg({
        '缺货金额': 'sum',
        '缺货数量': 'sum',
        '门店订货量': 'sum',
        '库存量': 'sum'
    })
    return summary.to_dict()

# 计算每个仓库的汇总数据
xa_summary = calculate_summary(xa_data)
gd_summary = calculate_summary(gd_data)
sz_summary = calculate_summary(sz_data)

# 为每个仓库创建一个数据结构，用于图表显示（不包括“可用天数”和“缺货天数”）
def create_chart_data(data):
    chart_data = []
    for column in ['缺货金额', '缺货数量', '门店订货量', '库存量']:
        series_data = {
            'name': column,
            'type': 'bar',
            'data': [data[column]]
        }
        chart_data.append(series_data)
    return chart_data

xa_chart_data = create_chart_data(xa_summary)
gd_chart_data = create_chart_data(gd_summary)
sz_chart_data = create_chart_data(sz_summary)

# 生成JS代码
js_code_chart1_corrected = f"""
var xaData = {json.dumps(xa_chart_data)};
var gdData = {json.dumps(gd_chart_data)};
var szData = {json.dumps(sz_chart_data)};

// 初始化图表
var chart1 = echarts.init(document.getElementById('chart1'));

// 定义更新图表的函数
function updateChart(data) {{
    var option1 = {{
        title: {{
            text: '仓库数据概览'
        }},
        tooltip: {{
            trigger: 'axis',
            axisPointer: {{ type: 'shadow' }}
        }},
        legend: {{
            data: ['缺货金额', '缺货数量', '门店订货量', '库存量']
        }},
        xAxis: {{
            type: 'category',
            data: ['数据']
        }},
        yAxis: {{
            type: 'value'
        }},
        series: data
    }};
    chart1.setOption(option1);
}}

// 绑定按钮事件
document.getElementById('btnA').addEventListener('click', function () {{
    updateChart(xaData);
}});
document.getElementById('btnB').addEventListener('click', function () {{
    updateChart(gdData);
}});
document.getElementById('btnC').addEventListener('click', function () {{
    updateChart(szData);
}});

// 默认加载第一个仓库的数据
updateChart(xaData);
"""

# 输出修正后的JS代码
print(js_code_chart1_corrected)
```

```python
import pandas as pd
import json

# 加载Excel文件
file_path = '../attachment/1596531496369353123.xlsx'
excel_data = pd.read_excel(file_path, sheet_name=None)

# 读取“供应商管理_3”工作簿
supplier_management_3_data = excel_data['供应商管理_3']
dates_3 = supplier_management_3_data['日期'].dt.strftime('%Y-%m-%d').tolist()
purchase_prices = supplier_management_3_data['采购单价'].tolist()

# 读取“供应商管理_1”工作簿
supplier_management_1_data = excel_data['供应商管理_1']
supplier_groups = supplier_management_1_data['联系人'].value_counts().reset_index()
supplier_groups.columns = ['联系人', '数量']

# 转换数据为适合绘图的格式
pie_chart_data = [{
    'name': row['联系人'],
    'value': row['数量']
} for index, row in supplier_groups.iterrows()]

# 生成动态图表（Chart1）的JS代码
js_code_chart1 = f"""
var data3 = {{
    '日期': {json.dumps(dates_3)},
    '采购单价': {json.dumps(purchase_prices)}
}};
var currentIndex = 0;
var updateInterval = 1000; // 更新间隔，例如1000毫秒（1秒）

// 初始化图表
var chart1 = echarts.init(document.getElementById('chart1'));
var option1 = {{
    title: {{
        text: '采购单价趋势'
    }},
    tooltip: {{
        trigger: 'axis'
    }},
    xAxis: {{
        type: 'category',
        data: []
    }},
    yAxis: {{
        type: 'value'
    }},
    series: [{{
        data: [],
        type: 'line'
    }}]
}};
chart1.setOption(option1);

// 定时添加新数据
setInterval(function() {{
    if (currentIndex < data3['日期'].length) {{
        var nextIndex = Math.min(currentIndex + 50, data3['日期'].length);
        for (var i = currentIndex; i < nextIndex; i++) {{
            option1.xAxis.data.push(data3['日期'][i]);
            option1.series[0].data.push(data3['采购单价'][i]);
        }}
        chart1.setOption(option1);
        currentIndex = nextIndex;
    }}
}}, updateInterval);
"""

# 生成饼图（Chart2）的JS代码
js_code_chart2 = f"""
var chart2 = echarts.init(document.getElementById('chart2'));
var option2 = {{
    title: {{
        text: '联系人占比'
    }},
    tooltip: {{
        trigger: 'item',
        formatter: '{{a}} <br/>{{b}}: {{c}} ({{d}}%)'
    }},
    series: [{{
        name: '联系人',
        type: 'pie',
        radius: '50%',
        data: {json.dumps(pie_chart_data)},
        emphasis: {{
            itemStyle: {{
                shadowBlur: 10,
                shadowOffsetX: 0,
                shadowColor: 'rgba(0, 0, 0, 0.5)'
            }}
        }}
    }}]
}};
chart2.setOption(option2);
"""

# 输出JS代码
print(js_code_chart1)
print(js_code_chart2)

```
